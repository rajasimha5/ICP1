```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('data.csv')
print(df.head())
print(df.describe())
print(df.isnull().sum())
print("\nAggregated Data (Calories and Pulse):")
aggregated = df[['Calories', 'Pulse']].agg(['min', 'max', 'count', 'mean'])
print(aggregated)
print("\nCalories between 500 and 1000:")
filtered1 = df[(df['Calories'] >= 500) & (df['Calories'] <= 1000)]
print(filtered1)
print("\nCalories > 500 and Pulse < 100:")
filtered2 = df[(df['Calories'] > 500) & (df['Pulse'] < 100)]
print(filtered2)
df_modified = df.drop(columns=['Maxpulse'])
print("\ndf_modified (without Maxpulse):")
print(df_modified.head())
df.drop(columns=['Maxpulse'], inplace=True)
df['Calories'].replace([np.inf, -np.inf], np.nan, inplace=True)
df['Calories'].fillna(df['Calories'].mean(), inplace=True)
df['Calories'] = df['Calories'].astype(int)
print("\nCalories column data type after conversion:")
print(df['Calories'].dtype)
plt.figure(figsize=(8, 5))
plt.scatter(df['Duration'], df['Calories'], alpha=0.6)
plt.title('Scatter Plot: Duration vs Calories')
plt.xlabel('Duration')
plt.ylabel('Calories')
plt.grid(True)
plt.tight_layout()
plt.show()
```
